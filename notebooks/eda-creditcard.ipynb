# Task 1: Exploratory Data Analysis - Credit Card Data
This notebook explores the credit card fraud dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import sys
sys.path.append(os.path.abspath('..'))
```

```python
credit_df = pd.read_csv('../data/raw/creditcard.csv')
print(f"Credit Card Data shape: {credit_df.shape}")
```

## 1. Class Distribution

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Class', data=credit_df)
plt.title('Credit Card Fraud Distribution')
plt.show()
print(credit_df['Class'].value_counts(normalize=True))
```

## 2. Feature Correlation

```python
plt.figure(figsize=(12, 10))
corr = credit_df.corr()
sns.heatmap(corr, cmap='coolwarm', annot=False)
plt.title('Correlation Heatmap')
plt.show()
```

## 3. Transaction Amount Analysis

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Class', y='Amount', data=credit_df[credit_df['Amount'] < 2000])
plt.title('Transaction Amount by Class (filtered < 2000)')
plt.show()
```
